При условии отсутствия отклонения от закона Дарси можно построить индикаторные кривые следующим образом:

```python
from datetime import date

title = 'Индикаторные кривые для скважины WELL08P'
x_axis = 'qн, м3/сут'
y_axis = 'Pwf, атм'

fig = go.Figure()
df = pd.read_csv('csv_folder/MODEL.0.csv', parse_dates=[0], index_col=[0])
j = 0
for i in df.index:
    if i == pd.to_datetime('2000-01-05 00:00:00') or i == pd.to_datetime('2005-12-30 00:00:00') or i == pd.to_datetime('2013-12-28 00:00:00'):
        fpr = df['FPR'][j]
        pwf = df['WBHP:WELL08P'][j]
        q = df['WOPR:WELL08P'][j]
        qfin = fpr/(fpr-pwf)*q
        x_list = [0, q, qfin]
        y_list = [fpr, pwf, 0]
        fig.add_trace(go.Scatter(
                x=x_list,
                y=y_list,
                mode='lines',
                name=f'{i}'))
    j += 1
fig.update_layout(title=go.layout.Title(text=title),
                       xaxis_title=x_axis,
                       yaxis_title=y_axis)
iplot(fig)
```
